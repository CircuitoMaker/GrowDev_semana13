```python
# 2 - Encontre o nome e o endereço dos clientes que fizeram um pedido em 2024.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
    SELECT DISTINCT C.Nome, C.Endereco
FROM Clientes C
JOIN Pedidos P ON C.ID_Cliente = P.ID_Cliente
WHERE EXTRACT(YEAR FROM P.Data_Pedido) = 2024;

""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 4 - Conte o número total de pedidos feitos no mês de julho de 2024.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT COUNT(*) AS Total_Pedidos
FROM Pedidos
WHERE EXTRACT(YEAR FROM Data_Pedido) = 2024 AND EXTRACT(MONTH FROM Data_Pedido) = 7;
""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 8 - Liste os pedidos com seu status e a quantidade de itens em cada pedido.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT P.ID_Pedido, P.Status, COUNT(IP.ID_Item) AS Quantidade_Itens
FROM Pedidos P
JOIN Itens_Pedido IP ON P.ID_Pedido = IP.ID_Pedido
GROUP BY P.ID_Pedido, P.Status;
""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 9 - Encontre os produtos que não foram vendidos em nenhum pedido.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT P.*
FROM Produtos P
LEFT JOIN Itens_Pedido IP ON P.ID_Produto = IP.ID_Produto
WHERE IP.ID_Produto IS NULL;

""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 10 - Liste o nome dos 10 maiores clientes e a quantidade de pedidos feitos por cada um. (por quantidade de pedidos)
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT C.Nome, COUNT(P.ID_Pedido) AS Quantidade_Pedidos
FROM Clientes C
JOIN Pedidos P ON C.ID_Cliente = P.ID_Cliente
GROUP BY C.Nome
ORDER BY Quantidade_Pedidos DESC
LIMIT 10;
""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 11 - Encontre o produto mais caro e o mais barato.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
-- Produto mais caro
SELECT *
FROM Produtos
ORDER BY Preco DESC
LIMIT 1;

-- Produto mais barato
SELECT *
FROM Produtos
ORDER BY Preco ASC
LIMIT 1;

""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 12 - Calcule a idade média dos clientes.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT AVG(EXTRACT(YEAR FROM AGE(Data_Nascimento))) AS Idade_Media
FROM Clientes;
""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 13 - Liste os clientes que possuem o mesmo sobrenome.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT *
FROM Clientes C1
WHERE EXISTS (
    SELECT 1
    FROM Clientes C2
    WHERE C1.ID_Cliente <> C2.ID_Cliente
    AND SPLIT_PART(C1.Nome, ' ', -1) = SPLIT_PART(C2.Nome, ' ', -1)
);

""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 14 - Encontre os pedidos feitos por clientes cujo nome começa com a letra 'A'.
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT P.*
FROM Pedidos P
JOIN Clientes C ON P.ID_Cliente = C.ID_Cliente
WHERE C.Nome LIKE 'A%';

""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```

```python
# 15 - Liste os produtos, sua categoria e uma indicação se estão em estoque ou esgotados (usando CASE):
import psycopg2

conn = psycopg2.connect(
    dbname="ecommerce",
    user="postgres",
    password="postgres2",
    host="localhost",
    port="5432"
)
cur = conn.cursor()

cur.execute("""
SELECT P.Nome, C.Nome AS Categoria,
       CASE
           WHEN P.Quantidade_Estoque > 0 THEN 'Em Estoque'
           ELSE 'Esgotado'
       END AS Status_Estoque
FROM Produtos P
JOIN Categorias C ON P.ID_Categoria = C.ID_Categoria;
""")
rows = cur.fetchall()
for row in rows:
    print(row)

cur.close()
conn.close()
```
